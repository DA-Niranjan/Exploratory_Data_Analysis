# DATA CLEANING - Application Data Set.

## Deleting null values

* Lets us check to find null values involved in the data set.

```python
null_percent = (application_data.isnull().mean()*100).round(2) 
null_percent[null_percent > 0].sort_values(ascending= False)
```

* Lets analyse the columns having null values more than 45 %

```python
application_data[null_percent[null_percent > 45].index]
```

```python
len(null_percent[null_percent > 45])
```

Observation : There are 49 columns with null values above 45 % and this features are related to appartment rented/owned by customer, which will not have impact on our analysis. Hence we can delete the columns.

```python
application_data = application_data[null_percent[null_percent < 45].index]
```

## Imputing the null values

* Occupation feature has 31% null values. But our analysis might have some relevance with this feature, hence replace Nan values with Unknown for analysis purpose.

```python
application_data["OCCUPATION_TYPE"].unique()
```

```python
application_data["OCCUPATION_TYPE"] = application_data["OCCUPATION_TYPE"].replace(np.NAN,"unknown")
application_data["OCCUPATION_TYPE"].value_counts(normalize=True)
```

```python
# For further analysis, lets make a list where we can use median values to replace null values - 

CON_COL = ["AMT_ANNUITY","AMT_GOODS_PRICE","AMT_REQ_CREDIT_BUREAU_HOUR","AMT_REQ_CREDIT_BUREAU_DAY","AMT_REQ_CREDIT_BUREAU_WEEK",
           "AMT_REQ_CREDIT_BUREAU_MON","AMT_REQ_CREDIT_BUREAU_QRT","AMT_REQ_CREDIT_BUREAU_YEAR","OBS_30_CNT_SOCIAL_CIRCLE","DEF_30_CNT_SOCIAL_CIRCLE",
           "OBS_60_CNT_SOCIAL_CIRCLE","DEF_60_CNT_SOCIAL_CIRCLE","DEF_60_CNT_SOCIAL_CIRCLE","DAYS_LAST_PHONE_CHANGE","EXT_SOURCE_2","EXT_SOURCE_3"]
for i in CON_COL:
  median = application_data[i].median()
  application_data[i].replace(np.NAN, median,inplace=True)

application_data.isna().mean()*100
```

## Identifying and deleting irrelevant columns

* Lets us check the relevance of other columns with respect to target values

```python
# Examine the document columns for any correlation with the target feature

document_col = [ 'FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_3','FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5', 'FLAG_DOCUMENT_6','FLAG_DOCUMENT_7', 
           'FLAG_DOCUMENT_8', 'FLAG_DOCUMENT_9','FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11', 'FLAG_DOCUMENT_12','FLAG_DOCUMENT_13',
           'FLAG_DOCUMENT_14', 'FLAG_DOCUMENT_15','FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17', 'FLAG_DOCUMENT_18',
           'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21']
df_doc= application_data[document_col+["TARGET"]]

df_doc.groupby(by="TARGET").mean()*100
```

Observation - From above we can infer that the submission rate of each type of document is same for both the target cases. Hence the features are not useful for analysis and can be deleted.

```python
relevance_check = ['FLAG_MOBIL', 'FLAG_EMP_PHONE', 'FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE','FLAG_PHONE', 'FLAG_EMAIL','EXT_SOURCE_2','EXT_SOURCE_3','TARGET']
plt.figure(figsize = [10,7])
sns.heatmap(application_data[relevance_check].corr(),annot=True,cmap="RdYlGn")
plt.show()
```

* Observations - The correlation of each contact details feature and EXT_SOURCE_X columns is low with the target varible. Hence this features are also irrelevant for analysis. Therefore we can delete the rows.

```python
# Group thr columns which we want to delete.
columns_to_delete = (relevance_check + document_col)
columns_to_delete.remove('TARGET')
len(columns_to_delete)
```

* There are 28 columns that are irrelevant and can be deleted from application data sheet.

```python
application_data.drop(columns_to_delete,axis =1,inplace=True)
```

## MultiVariate Analysis

```python
"""
Define a function for developing tables and chart for finding rate of clients without payment difficulties in various categorical
features with respect to NAME_CONTRACT_STATUS 
"""


def final_review_plot(x):
  def repayer(a):
    return (1-a.mean()) *100

  df = pd.pivot_table(final_review, index =x ,columns ="NAME_CONTRACT_STATUS", values = "TARGET",aggfunc = repayer)
  print(df,"\n")

  df.plot.bar(figsize = [24,8])
  plt.legend(bbox_to_anchor=(1.02, 0), loc="lower left", borderaxespad=0)
  plt.ylabel("Percentage %")
  plt.title(f"Percentage of Repayers in {x} and every Status\n", fontdict={"fontsize": 20, "color":"black", "fontweight":5})
  plt.show()
```

Using function plot graphs for various feature

```python
final_review_plot("CODE_GENDER")
```

```python
final_review_plot("NAME_EDUCATION_TYPE")
```
